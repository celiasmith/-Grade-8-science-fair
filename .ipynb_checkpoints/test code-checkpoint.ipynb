```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x,step = np.linspace(-10.0, 10.0, num=200,retstep=True)
```

```python
m =5*x*x*x
b =0
y = m*x+b
w=x
derps = x*x*x*x*x
```

```python
plt.plot(x,y,'k')
plt.plot(x,w,'r')
plt.plot(x,derps,'g')
grid('on')
plt.plot(x[1:],np.diff(derps)/step,'b')
```
