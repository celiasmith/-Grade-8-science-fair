```python
%pylab inline
import matplotlib.pylab as plt
import numpy as np
import time
import datetime
from IPython import display
import pygame
import serial
ser = serial.Serial('/dev/tty.usbmodem1421', 115200)
#ser.write(' ')
# Define some colors
BLACK    = (   0,   0,   0)
WHITE    = ( 255, 255, 255)
GREEN    = (   0, 255,   0)
RED      = ( 255,   0,   0)
BLUE     = (   0,   0, 255)

def circdist(theta, TargetTheta):
    theta = np.fmod(theta,np.pi*2)
    if theta < 0:
        theta = theta + 2*np.pi
    error = TargetTheta-theta
    if error > np.pi:
            error = error - 2*np.pi
    elif error < -np.pi:
            error = error + 2*np.pi
    return error 

# Setup
pygame.init()
   
# Set the width and height of the screen [width,height]
size = [500, 500]
screen = pygame.display.set_mode(size)
   
#Loop until the user clicks the close button.
done = False
  
# Used to manage how fast the screen updates
clock = pygame.time.Clock()
 
# Hide the mouse cursor
pygame.mouse.set_visible(0)
#############
# Constants #
#############
xspacer = 250
yspacer = 250
pixelspm = 3000    #pixels per meter
font = pygame.font.Font(None, 20)

fricoeff = 0.8    #friction strength
r=0.05   #measured in meters
g=0     #the acceleration due to the gravity of earth in m/s^2
theta=np.pi*3/2+3#starting theta
run_time = 30. #seconds to simulate
dt = 1./30 #seconds (30 fps)

score = 0
error = 0
TargetTheta = np.pi #np.random.rand()*np.pi*2
xtarget = np.sin(TargetTheta)*r*pixelspm
ytarget = np.cos(TargetTheta)*r*pixelspm
dtheta = 0
ddtheta = 0
Time = run_time/dt
touched = True
showtime = 0
led_pos = [320,400,457,525,593,655,700,767,830,890,945,1000,32,96,172,255]
led_order = [1,14,15,12,13,10,11,8,9,6,7,4,5,2,3,16]
#plt.figure(figsize=(12,12))
#plt.axes().set_aspect('equal')
##############################COMBOS strength kp  kd                                   XXXXXXXXXXXXXXXXXXXXXXXXX

##############################        2     0.5  0.2
##############################        pi    0.5  0.1
kp = 2.0
kd = 0.3
savedforce = []
#############
# Main Loop #
#############
time.sleep(2) #Let serial connection initialize
# Pick the first target
index = 4#np.ceil(float(np.random.rand(1)*16-1))
target = led_order[int(index)]
targetposition = led_pos[int(index)]
ser.flushOutput()
ser.write(chr(int(target)))
ser.flushInput()

position = 0
controlled = True  
tmpstr=" "
while not done:

    #for t in range(0,np.floor(run_time/dt).astype(np.int64)):
    for event in pygame.event.get(): # User did something
        if event.type == pygame.QUIT: 
            done = True    
    try:
        if ser.inWaiting() > 3:
            old_pos = position
            position = ser.readline().split()
            if position != []:
                position = int(position[0])  
            else:
                position = old_pos
                            
            ser.flushInput()
    except serial.serialutil.SerialException:
        print 'Serial Error. Continuing...'
        
    if controlled == True:
        ####################
        #    Controller    #
        ####################
        theta = 2*np.pi/1024*position
        TargetTheta = 2*np.pi/1024*targetposition
        lasterror = error
        error = -circdist(theta, TargetTheta)
        derror = (error - lasterror)/dt
        force = (error * kp) + (derror * kd)
    
        if force > 1:
            force = 1.
        elif force < -1:
            force = -1.
        force = force/50

    else:
        ####################
        #  Joystick Input  #
        ####################

        joystick = pygame.joystick.Joystick(0)
        joystick.init()

        axis2 = joystick.get_axis(2)
        if axis2 < 0.01 and axis2 > -0.01:                  #add deadzone
            axis2 = 0
        force = axis2/50
        
    #########
    #arduino#
    #########
    friction = dtheta*fricoeff
    force = (force*15000)
    ddtheta = force - friction
    #if dtheta > 300 and ddtheta>0: ddtheta = 0
    #elif dtheta < -300 and ddtheta<0: ddtheta = 0
    #dtheta = dtheta + ddtheta
    
    #print ddtheta
    if -10 < dtheta < 10:
        tmpstr = " "
        ser.write(tmpstr)        
    
    if ddtheta > 0.02:
        tmpstr = int(round(ddtheta/10))*","
        dtheta += int(round(ddtheta/10))
        ser.write(tmpstr)
    
    elif ddtheta < -0.02:
        tmpstr = -int(round(ddtheta/10))*"."
        dtheta += int(round(ddtheta/10))
        ser.write(tmpstr)
        
    if dtheta > 200:
        dtheta = 200
    if dtheta < -200:
        dtheta = -200
   
    pygame.display.set_caption("showtime={:.2f}".format(showtime))
    screen.fill(WHITE)
    
    ######################
    # Target Testing co. #
    ######################
    if position < targetposition+20 and position > targetposition-20 and dtheta > -30 and dtheta < 30:
        print 'Got Target!'
        old_ind = index
        index = np.ceil(float(np.random.rand(1)*16-1))
        if old_ind != index:
            target = led_order[int(index)]
            targetposition = led_pos[int(index)]
            ser.flushOutput()
            time.sleep(.05)
            ser.write(chr(int(target)))
            score += 1

    ########
    # Time #
    ########
    Time = Time - 1
    showtime = Time*dt
    if Time < 0:
        print score
        done = True
        
    ####################
    # Show Results     #
    ####################
    screen.fill(WHITE)
    
    force_info = "dtheta: {:f}, ddtheta: {:f}, error: {:}, force: {:f}".format(dtheta, ddtheta,error,force)
    
    bitmap = font.render(force_info, True, BLACK)
    screen.blit(bitmap,(10,30))

    scorestring = "score: {:d}".format(score)
    bitmap = font.render(scorestring, True, BLACK)
    screen.blit(bitmap,(10,10))

    tmp = "target: {:}, position: {:}, targetposition: {:}".format(int(target),theta,TargetTheta)
    bitmap = font.render(tmp, True, BLACK)
    screen.blit(bitmap,(10,50))    
    
    display.clear_output(wait=True)
    #display.display(plt.gcf())
    #plt.clf()
    #plt.show()
    #time.sleep(0.1)    
    pygame.display.flip()
    
    # Limit to 60 frames per second
    #clock.tick(30)
    savedforce.append(force)

# Close the window and quit.
pygame.quit()

print score
plt.plot(savedforce)
with open("RESTART THE STUPID THING FILE.txt", "a") as myfile:
   myfile.write("{0}, Computer,1,0.3, {1:d}\n".format(datetime.datetime.now(), score))
```
