```python
%matplotlib inline
import matplotlib.pylab as plt
import datetime
import numpy as np
import time
from IPython import display
import pygame

# Define some colors
BLACK    = (   0,   0,   0)
WHITE    = ( 255, 255, 255)
GREEN    = (   0, 255,   0)
RED      = ( 255,   0,   0)
BLUE     = (   0,   0, 255)
  
def circdist(theta,TargetTheta):
    theta = np.fmod(theta,np.pi*2)
    if theta < 0:
        theta = theta + 2*np.pi
    error = TargetTheta-theta
    if error > np.pi:
            error = error - 2*np.pi
    elif error < -np.pi:
            error = error + 2*np.pi
    return error 

# Setup
pygame.init()
   
# Set the width and height of the screen [width,height]
size = [500, 500]
screen = pygame.display.set_mode(size)
   
#Loop until the user clicks the close button.
done = False
  
# Used to manage how fast the screen updates
clock = pygame.time.Clock()
 
# Hide the mouse cursor
pygame.mouse.set_visible(0)

#############
# Constants #
#############
xspacer = 250
yspacer = 250
pixelspm = 3000    #pixels per meter
font = pygame.font.Font(None, 20)

fricoeff = 0.08    #friction strength
r=0.05   #measured in meters
g=0     #the acceleration due to the gravity of earth in m/s^2
run_time = 30. #seconds to simulate
dt = 1./30 #seconds (10ms)
for i in np.arange(0,3):
    for kd in np.arange(0.2,3.6,0.3):
        for kp in np.arange(26,36,3):
            theta=np.pi*3/2+3#starting theta
            error = 0
            score=0
            TargetTheta = np.pi #np.random.rand()*np.pi*2
            xtarget = np.sin(TargetTheta)*r*pixelspm
            ytarget = np.cos(TargetTheta)*r*pixelspm
            dtheta = 0
            time = run_time/dt
            touched = True
            showtime = 0
            #plt.figure(figsize=(12,12))
            #plt.axes().set_aspect('equal')
            #strength = 17.5
            ##############################COMBOS strength kp  ki  kd 
            ##############################        17.5   0.5  1   0.05
            ##############################
            #kp = 0.5*strength
            #ki = 1
            #kd = 0.05*strength
            savedforce = []

            #############
            # Main Loop #
            ######################################################################################XXXXXXXXXXXXXX
            controlled = True

            starter = False
            while starter == False:
                for event in pygame.event.get(): # User did something
                    if event.type == pygame.QUIT: 
                        done = True    
                joystick = pygame.joystick.Joystick(0)
                joystick.init()
                a = joystick.get_button(2)
                starter = True


                screen.fill(WHITE)
                pygame.display.flip()
                clock.tick(30)

            while not done:
                #for t in range(0,np.floor(run_time/dt).astype(np.int64)):
                for event in pygame.event.get(): # User did something
                    if event.type == pygame.QUIT: 
                        done = True    

                if controlled == True:
                    ####################
                    #    Controller    #
                    ####################
                    lasterror = error
                    error = circdist(theta, TargetTheta)
                    derror = (error - lasterror)/dt
                    force = (error * kp) + (derror * kd)



                    if force > 1:
                        force = 1.
                    elif force < -1:
                        force = -1.
                    force = force/50

                else:
                    ####################
                    #  Joystick Input  #
                    ####################

                    joystick = pygame.joystick.Joystick(0)
                    joystick.init()
                    axis2 = joystick.get_axis(2)
                    if axis2 < 0.01 and axis2 > -0.01:                  #add deadzone
                        axis2 = 0
                    force = axis2/50

                    #axis1 = joystick.get_axis(1)
                    #r=r + axis1/100

                ####################
                # Motion equations #
                ####################

                friction = dtheta*fricoeff
                ddtheta=-g/r*np.sin(theta)*dt**2+force-friction
                dtheta=dtheta+ddtheta
                theta=theta+dtheta

                x=np.sin(theta)*r*pixelspm
                y=np.cos(theta)*r*pixelspm

                pygame.display.set_caption("showtime={:.2f}".format(showtime))
                #pygame.display.set_caption("theta={:.2f}".format(theta))
                screen.fill(WHITE)

                ######################
                # Target Testing co. #
                ######################
                if (y < ytarget+10 and y > ytarget-10) and (x < xtarget+10 and x > xtarget-10) \
                            and (dtheta < 0.02 and dtheta > -0.02):
                    touched = True
                    score = score+1
                if touched:
                    touched = False
                    while ((y < ytarget+50 and y > ytarget-50) and (x < xtarget+50 and x > xtarget-50)):
                        TargetTheta = np.random.rand()*np.pi*2
                        xtarget = np.sin(TargetTheta)*r*pixelspm
                        ytarget = np.cos(TargetTheta)*r*pixelspm



                #TargetTheta = TargetTheta+axis2/3

                ########
                # Time #
                ########
                time = time - 1
                showtime = time*dt
                if time < 0:
                    print score
                    done = True
                ####################
                # Plotting Results #
                ####################
                #scorestring="error: %1.2f theta: %1.2f Fmodtheta: %1.2f TargetTheta %1.2f" %(error,theta,np.fmod(theta,np.pi*2),TargetTheta)
                scorestring = "score: {:d}".format(score)
                bitmap = font.render(scorestring, True, BLACK)
                screen.blit(bitmap,[10,10])
                pygame.draw.ellipse(screen, BLUE, [xtarget+xspacer, -ytarget+yspacer, 50, 50], 0)
                pygame.draw.line(screen, BLACK, [x+xspacer+25, -y+yspacer+25], [xspacer+15, yspacer+15], 4)
                pygame.draw.ellipse(screen, GREEN, [x+xspacer, -y+yspacer, 50, 50], 0)
                pygame.draw.ellipse(screen, RED, [xspacer, yspacer, 30, 30], 0)

                display.clear_output(wait=True)
                #display.display(plt.gcf())
                #plt.clf()
                #plt.show()
                #time.sleep(0.1)    
                pygame.display.flip()

                # Limit to 30 frames per second
                clock.tick(30)
                savedforce.append(force)

            # Close the window and quit.
            # If you forget this line, the program will 'hang'
            # on exit if running from IDLE.

            print score
            score
            plt.plot(savedforce)
            with open("More Controllers.txt", "a") as myfile:
                myfile.write("{0},Computer,{1},{2},{3:d}\n".format(datetime.datetime.now(),kp,kd, score))
            done = False

pygame.quit()
```
